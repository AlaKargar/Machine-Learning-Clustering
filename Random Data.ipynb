```python
import sklearn.cluster
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

x, y= make_blobs(n_samples= 300, centers= 4, cluster_std= 0.6, random_state= 42)
k_means= KMeans(n_clusters= 4, init= 'k-means++', random_state= 0)
y_pred= k_means.fit_predict(x)
labels= k_means.labels_
plt.scatter(x[:, 0], x[:, 1], c= y_pred, cmap= 'viridis')
plt.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:, 1], s=200, c='red', marker= 'x')

sill= silhouette_score(x, labels)
dbi= davies_bouldin_score(x, labels)
ch= calinski_harabasz_score(x, labels)
print("silhouette score:", sill)
print("Davies_bouldin_score:", dbi)
print("Calinski_harabasz_score", ch)
```

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
z= linkage(x, method='ward')
dendrogram(z)
plt.show()

agg= AgglomerativeClustering(n_clusters= 4, linkage= 'ward')
y_pred= agg.fit_predict(x)
labels2= agg.labels_
sill= silhouette_score(x, labels2)
dbi= davies_bouldin_score(x, labels2)
ch= calinski_harabasz_score(x, labels2)

print("Silhouette Score:", sill)
print("Davies-Bouldin Score:", dbi)
print("Calinski-Harabasz Score:", ch)
```

```python
from sklearn.cluster import DBSCAN

db= DBSCAN(eps=1, min_samples=5)
y_pred= db.fit_predict(x)
labels3= db.labels_
plt.scatter(x[:,0], x[:, 1], c= y_pred, cmap= 'plasma')
plt.show()

sill= silhouette_score(x, labels3)
dbi= davies_bouldin_score(x, labels3)
ch= calinski_harabasz_score(x, labels3)

print("Silhouette Score:", sill)
print("Davies-Bouldin Score:", dbi)
print("Calinski-Harabasz Score:", ch)
```
