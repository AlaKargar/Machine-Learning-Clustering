```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN, AgglomerativeClustering, KMeans
from sklearn.mixture import GaussianMixture
```

Data Scaling 

```python
iris = load_iris()
x = iris.data
y_true = iris.target  

scaler= StandardScaler()
x_scaled= scaler.fit_transform(x)
```

Fitting Model

```python
kmeans= KMeans(n_clusters=3, init='k-means++', random_state= 42)
heirarchy= AgglomerativeClustering(n_clusters=3, linkage= 'ward')
dbscan= DBSCAN(eps=1, min_samples=5)
gmm= GaussianMixture(n_components= 4, covariance_type='full', random_state= 42)
```

```python
label_kmeans= kmeans.fit_predict(x_scaled)
label_heirarchy= heirarchy.fit_predict(x_scaled)
label_dbscan= dbscan.fit_predict(x_scaled)
label_gmm= gmm.fit_predict(x_scaled)
```

KMeans Scores

```python
print('Silhoutte: ', silhouette_score(x, label_kmeans))
print('davies bouldin:', davies_bouldin_score(x_scaled, label_kmeans))
print('calinski harabasz:', calinski_harabasz_score(x_scaled, label_kmeans))
```

DBScan Scores

```python
print('Silhoutte: ', silhouette_score(x, label_dbscan))
print('davies bouldin:', davies_bouldin_score(x_scaled, label_dbscan))
print('calinski harabasz:', calinski_harabasz_score(x_scaled, label_dbscan))
```

Heirarchy Scores

```python
print('Silhoutte: ', silhouette_score(x, label_heirarchy))
print('davies bouldin:', davies_bouldin_score(x_scaled, label_heirarchy))
print('calinski harabasz:', calinski_harabasz_score(x_scaled, label_heirarchy))
```

GaussianMixture Score

```python
print('Silhoutte: ', silhouette_score(x, label_gmm))
print('davies bouldin:', davies_bouldin_score(x_scaled, label_gmm))
print('calinski harabasz:', calinski_harabasz_score(x_scaled, label_gmm))
```

```python
k_means = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_pred = k_means.fit_predict(x)

sill = silhouette_score(x, y_pred)
dbi = davies_bouldin_score(x, y_pred)
ch = calinski_harabasz_score(x, y_pred)

print("Silhouette Score:", sill)
print("Davies-Bouldin Score:", dbi)
print("Calinski-Harabasz Score:", ch)

plt.scatter(x[:, 0], x[:, 1], c=y_pred, cmap='viridis')
plt.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:, 1], 
            s=200, c='red', marker='x', label='Centers')
plt.title('K-Means Clustering on Iris Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
